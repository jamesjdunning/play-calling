# NFL Play Calling Machine Learning Challenge: Part 2

#### Introduction:

Hello! In this notebook I will go through the process of building a Random Forest classifier to predict NFL play calls. Given that my dataset contains more than 600,000 observations, I found the algorithms to be very computationally intensive. For that reason, I have made some simplifications with feature selection and hyperparameter tuning.

See Part 1 of my analysis [here](https://nbviewer.jupyter.org/github/jamesjdunning/play-calling/blob/master/notebooks/part-1.ipynb).

```python
#Pandas is the library of choice for managing my datasets
import pandas as pd
import numpy as np

#Read in excel file as a pandas dataframe
data = pd.read_excel('..//data/data.xlsx', sheet_name = 'data')
```

```python
train = data.copy()
print('There are {} rows and {} columns in the dataset.'.format(train.shape[0], train.shape[1]))
```

#### Filling in missing values

This was covered in the first notebook, but it's important to fill in missing values before fitting a classifier. Here I fill in missing values for three columns: temperature, wind, and last_play_gain.

```python
# Set missing values to median outdoors values
median_temp = train[train.venue == 'outdoors']['temperature'].median()
median_wind = train[train.venue == 'outdoors']['wind'].median()

# Fill null values
train.temperature.fillna(median_temp, inplace = True)
train.wind.fillna(median_wind, inplace = True)
train.last_play_gain.fillna(0, inplace = True)
```

#### Partition numeric and categorical features

When preprocessing the dataset, numeric and categorical features require different tasks. Numeric features can be scaled, while categorical features can be encoded as dummy variables. For that reason I am splitting the features up. Additionally, I am storing the target feature as y and then dropping it from the training data.

```python
# Differentiate numerical features (minus the target) and categorical features
categorical_features = train.select_dtypes(include = ['object']).columns
categorical_features = categorical_features.drop(['head_coach', 'off_coordinator', 'play_type'])

numerical_features = train.select_dtypes(exclude = ['object']).columns

print("Numerical features : " + str(len(numerical_features)))
print("Categorical features : " + str(len(categorical_features)))

#Split numerical, categorical, and target features
train_num = train[numerical_features]
train_cat = train[categorical_features]
y = train.play_type
```

```python
train_cat = pd.get_dummies(train_cat)
print("Categorical features : " + str(len(train_cat.columns)))
```

```python
# Join categorical and numerical features
train = pd.concat([train_num, train_cat], axis = 1)
print("New number of features : " + str(train.shape[1]))
```

#### Split the dataset into training and test data

I will train the classifier on the training data, and measure its accuracy on the test data. Since I have a large dataset, I've manually set the test data size to be larger than default, at 35%.

```python
#Load scikit's model evaluation and preprocessing library
from sklearn.model_selection import train_test_split
```

```python
# Partition the dataset in train + validation sets
X_train, X_test, y_train, y_test = train_test_split(train, y, test_size = 0.35, random_state = 0)

# Print shape of different data sets
print("X_train : " + str(X_train.shape))
print("X_test : " + str(X_test.shape))
print("y_train : " + str(y_train.shape))
print("y_test : " + str(y_test.shape))
```

#### Unsupervised transformer: Principal Component Analysis and Manfiold Learning

In order to visualize this high dimension training dataset, I will perform some dimensionality reduction with Principal Component Analysis and Manfiold Learning with t-SNE. Due to memory restrictions, I am selecting only 10,000 observations from the dataset with manifold learning.

```python
from sklearn.decomposition import PCA
```

```python
# Create PCA object and fit to dataset
pca = PCA(n_components=2)
pca.fit(train)

print(train.shape)
X_pca = pca.transform(train)
X_pca.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

y_num = pd.Categorical(y).codes
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=plt.cm.Vega10(y_num))
```

The PCA visualization does not provide any intuitive results for me, other than the target feature classes are not as distinct as I would have hoped.

```python
from sklearn.manifold import Isomap
```

```python
isomap = Isomap()

X_isomap = isomap.fit_transform(X_test.sample(10000))
```

```python
plt.figure()
plt.scatter(X_isomap[:, 0], X_isomap[:, 1], c=plt.cm.Vega10(y_num))
```

This visualization is better than the PCA result. However, it confirms what I suspected from the prior chart, that the classes are not as distinct as desired.

#### Scaling numerical data

When building linear models, it is a good idea to scale your numerical data. In this case I am building a Random Forest Classifier, which is not linear, so I do not need to scale my numerical data. However, I will do so anyways as an example, since it will not affect my results.

```python
from sklearn.preprocessing import StandardScaler

# Standardize numerical features
stdSc = StandardScaler()
X_train.loc[:, numerical_features] = stdSc.fit_transform(X_train.loc[:, numerical_features])
X_test.loc[:, numerical_features] = stdSc.transform(X_test.loc[:, numerical_features])
```

#### Creating my classifier

One important parameter to determine when creating a Random Forest Classifier is the number of trees, or n_estimators, to build. Here I plot the effect of n_estimators on the classifier's accuracy. While using all other default parameters, the accuracy effect of n_estimators appears to plateau at 20. I'll use that going forward.

```python
from sklearn.ensemble import RandomForestClassifier

train_scores = []
test_scores = []

clf = RandomForestClassifier(warm_start=True)
estimator_range = range(1, 100, 10)

for n_estimators in estimator_range:
    clf.n_estimators = n_estimators
    clf.fit(X_train, y_train)
    train_scores.append(clf.score(X_train, y_train))
    test_scores.append(clf.score(X_test, y_test))
```

```python
plt.plot(estimator_range, test_scores, label="test scores")
plt.plot(estimator_range, train_scores, label="train scores")
plt.ylabel("accuracy")
plt.xlabel("n_estimators")
plt.legend()
```

Next I'll look to tune my other hyperparameters and perform some cross-validation. I can do both with GridSearchCV, where I can perform a loop over each parameter setting and find the combination of paramaters that produce the highest accuracy score. Due to CPU restrictions, I am performing the grid search over a randomly select subset of 100,000 observations.

```python
from sklearn.model_selection import GridSearchCV 
```

```python
clf = RandomForestClassifier()

param_grid = {'n_estimators' : [20],
              'max_features' : ['sqrt', 15, 30],
              'min_samples_leaf' : [10, 25, 50],
              'n_jobs' : [-1]}

grid_search = GridSearchCV(clf, param_grid, cv = 5)

X_train_sample = X_train.sample(100000)
y_train_sample = y_train[X_train_sample.index]
X_test_sample = X_test.sample(100000)
y_test_sample = y_test[X_test_sample.index]

grid_search.fit(X_train_sample, y_train_sample)

#Print training and test set scores
print('Train set score: {:.3f}'.format(grid_search.score(X_train_sample, y_train_sample)))
print('Test set score: {:.3f}'.format(grid_search.score(X_test_sample, y_test_sample)))

#Print parameter info
print('Best parameters: {}'.format(grid_search.best_params_))
print('Best parameters: {}'.format(grid_search.best_score_))
```

Given these parameters, I'll re-create my classifier, fitting it on the entire training dataset. With these parameters, the classifiers accuracy is 71.5% on the test dataset. While this is good, I am little disappointed it did not come closer to 75%.

```python
#Fit classifier
clf = RandomForestClassifier(n_estimators = 20, max_features = 15, min_samples_leaf = 25, n_jobs = -1)
clf.fit(X_train, y_train)

#Print training and test set scores
print('Train set score: {:.3f}'.format(grid_search.score(X_train, y_train)))
print('Test set score: {:.3f}'.format(grid_search.score(X_test, y_test)))

#Print parameter info
print('Best parameters: {}'.format(grid_search.best_params_))
print('Best parameters: {}'.format(grid_search.best_score_))
```

#### Confusion Matrix

Below I'll plot a confusion matrix using code borrowed from sklearn's [documentation](http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html). With this chart I can gain a better understanding of which classes I am correctly predicting and those which I am not.

The classifier correctly predicts run nearly 73% of the time, with 27% of the predictions incorrectly labeled as Run. Similarly, the classifier correctly predicts run 64% of the time, with 34% of the predictions incorrectly labeled as Pass. I knew this would be the main challenge when I started this project, correctly choosing between Run and Pass. There remains some work to be done here, maybe via feature engineering related to 1st & 10 and short yardage downs.

On the positive side, my classifier is doing a good job predicting punts, correctly predicting punt over 98% of the time. On a similar note, the classifier correctly predicts Field goal 86% percent of the time.

On the negative side, my classifier does not correctly predict Spike once. Perhaps I'll need to engineer new features related to time_remaining to assist my algorithm. 

```python
from sklearn.metrics import confusion_matrix
import itertools

y_pred = grid_search.predict(X_test)
class_names = ['Field goal', 'Kneel', 'Pass', 'Punt', 'Run', 'Spike']

# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
np.set_printoptions(precision=2)
np.set_printoptions(suppress=True)

#Chart confusion matrix
fig, ax = plt.subplots(figsize=(10,6))
plt.imshow(cnf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Confusion matrix")
plt.colorbar()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names, rotation=45)
plt.yticks(tick_marks, class_names)

sums = (cnf_matrix.sum(axis = 1))
print(cnf_matrix / sums.reshape((6,1)) * 100)

thresh = cnf_matrix.max() / 2.
for i, j in itertools.product(range(cnf_matrix.shape[0]), range(cnf_matrix.shape[1])):
    plt.text(j, i, cnf_matrix[i, j],
             horizontalalignment="center",
             color="white" if cnf_matrix[i, j] > thresh else "black")

plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')

plt.show()
```
